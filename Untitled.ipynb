```python
import requests, os, urllib.parse, time

def download_cover(title, author=None):
    """
    Downloads the book cover using OpenLibrary API and saves it locally.
    Works even offline or if placeholder site fails.
    """
    # Folder to save covers
    os.makedirs("book_covers", exist_ok=True)

    # Clean filename
    safe_title = "".join(c if c.isalnum() else "_" for c in title)[:100]
    filename = f"book_covers/{safe_title}.jpg"

    # If already downloaded, skip
    if os.path.exists(filename):
        return filename

    # 1️⃣ Try OpenLibrary API
    try:
        search_url = "https://openlibrary.org/search.json"
        params = {"title": title}
        if author:
            params["author"] = author

        res = requests.get(search_url, params=params, timeout=20)
        data = res.json()

        if data.get("docs"):
            for doc in data["docs"]:
                if "cover_i" in doc:
                    cover_id = doc["cover_i"]
                    cover_url = f"https://covers.openlibrary.org/b/id/{cover_id}-L.jpg"
                    img = requests.get(cover_url, timeout=20)
                    if img.status_code == 200:
                        with open(filename, "wb") as f:
                            f.write(img.content)
                        return filename
    except Exception as e:
        print(f"⚠️ OpenLibrary fetch failed for '{title}': {e}")

    # 2️⃣ Try fallback placeholder safely
    try:
        placeholder = f"https://via.placeholder.com/200x300.png?text={urllib.parse.quote_plus(title[:25])}"
        r = requests.get(placeholder, timeout=10)
        with open(filename, "wb") as f:
            f.write(r.content)
        return filename
    except Exception as e:
        print(f"⚠️ Placeholder download failed for '{title}': {e}")

    # 3️⃣ Last resort: create an empty file
    with open(filename, "wb") as f:
        pass
    return filename
```

```python
from tqdm import tqdm

books_to_download = [
    "Carter Beats The Devil",
    "The Halfling'S Gem",
    "Don'T Judge A Girl By Her Cover",
    "The Other Side Of The Island",
    " King Arthur And His Knights Of The Round Table",
    " The Declaration",
]

for book in tqdm(books_to_download):
    path = download_cover(book)
    print(f"✅ {book} → {path}")
```
